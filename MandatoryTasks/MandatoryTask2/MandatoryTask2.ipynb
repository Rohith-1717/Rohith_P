```python
import pandas as pd
import torch
from torch import nn
from sklearn.metrics import mean_squared_error, r2_score

ds = "/kaggle/input/regression/train_data.csv"
testing = "/kaggle/input/regression/test_data.csv"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class LR(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(2, 1)

    def forward(self, x):
        return self.linear(x)

bs = 100000
lrate = 0.001

m1 = LR().to(device)
msl = nn.MSELoss()
adam = torch.optim.Adam(m1.parameters(), lr=lrate)
epochs = 10




for epoch in range(1,epochs+1):
    tl = 0
    batches = 0

    for chunk in pd.read_csv(ds, chunksize=bs):
        X = torch.tensor(chunk[['w', 'x']].values, dtype=torch.float32).to(device)
        y = torch.tensor(chunk['y'].values, dtype=torch.float32).view(-1, 1).to(device)

        adam.zero_grad()
        preds = m1(X)
        loss = msl(preds, y)
        loss.backward()
        adam.step()

        tl+=loss.item()
        batches+=1

    al = tl/batches
    print(f"Curent {epoch} out of {epochs}: Loss is {al:.10f}")

print("Training is over")
print("Testing is startiing")


ytest = []
ymodel = []

ytest = []
ymodel = []

for chunk in pd.read_csv(testing, chunksize=bs):
    X = torch.tensor(chunk[['w','x']].values, dtype=torch.float32).to(device)

    with torch.no_grad():
        preds = m1(X).cpu().numpy()
    ymodel.extend(preds)

print("Testing is over")
print("First 10 predictions:", ymodel[:10])


```
